```python

def saveDataframe(df, filename):
    df.to_pickle("./"+filename)

def loadDataframe(filename):
    df = pd.read_pickle("./"+filename)
    return df
```

```python
import pandas as pd
symbols = ['GOOGL','AAPL']
volumes = [10,12]
prices  = [1023,827]
dates   = [pd.Timestamp('2018-03-10 09:00'),pd.Timestamp('2018-03-10 09:00')]
actions = ['Buy','Buy']
portfolio = pd.DataFrame(
    {'symbol': symbols,
     'volume': volumes,
     'price' : prices,
     'date'  :dates,
     'action':actions
    })
portfolio['dt']=portfolio['date'].dt.date
saveDataframe(portfolio, 'portfolio.pkl')
```

```python
portfolio = loadDataframe('portfolio.pkl')
portfolio
```

```python
def register(symbols,volumes, prices,dates,actions):
    pf = pd.DataFrame(
    {'symbol': symbols,
     'volume': volumes,
     'price' : prices,
     'date'  :dates,
     'action':actions
    })
    pf['dt']=pf['date'].dt.date
    df = loadDataframe('portfolio.pkl')
    df2=df.append(pf)
    saveDataframe(df2, 'portfolio.pkl')

register(['GOOGL'],[10],[1023],[pd.Timestamp('2018-03-10 09:00')],['Buy'])
loadDataframe('portfolio.pkl')


def getTradingDays(symbol):
    data, errorCode = sh.downloadSymbolHistory(symbol,'2999-12-31', False)
    return data.index.tolist()
#getTradingDays('NLFSK.CPH')
#data
```

```python
# Downloading closing prices
from pandasql import sqldf
uniqueSymbols = sqldf('select symbol from portfolio order by symbol')['symbol']
for symbol in uniqueSymbols:
    symbolactions = sqldf('select date as Date, volume, price from portfolio where symbol="'+symbol+'"')
    data,errorCode=sh.downloadSymbolHistory(symbol, maxDate='2999-12-31', download=False)
    portfolioDays = pd.merge(data, symbolactions, on='Date', how='outer').sort_values(by='Date')
    portfolioDays["volume"].fillna(0, inplace = True)
    portfolioDays['rollingVolume'] = portfolioDays['volume'].cumsum(axis = 0)
portfolioDays.query('Date > "2018-03-01"')
    
```
